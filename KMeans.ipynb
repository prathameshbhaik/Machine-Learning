```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('wine.data')
```

```python
data.columns = ['class','Alcohol','Malic acid','Ash','Alcalinity of ash','Magnesium','Total phenols',
                'Flavanoids','Nonflavanoid phenols','Proanthocyanins','Color intensity','Hue',
                'OD280/OD315 of diluted wines','Proline']

data.head()
```

```python
data.isnull().sum()
```

```python
X = data.iloc[:,1:]
```

```python
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
scaled_data = ss.fit_transform(X)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
x_pca = pca.fit_transform(scaled_data)
```

```python
print(scaled_data.shape)   #before
print(x_pca.shape)         #after
```

```python
X_PCA = pd.DataFrame(x_pca,columns=["P1","P2"])
X_PCA.head()
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=3)
y_label = km.fit_predict(X_PCA)
y_label
```

```python
km.cluster_centers_
```

```python
plt.scatter(X_PCA['P1'],X_PCA['P2'],c=y_label)
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],marker='*',color='red')
plt.show()
```

```python
SSE = []
for cluster in range(1,20):
    kmeans = KMeans(n_clusters=cluster)
    kmeans.fit(scaled_data)
    SSE.append(kmeans.inertia_)
```

```python
plt.plot(range(1,20),SSE,marker='o')
plt.xlabel('No. of clusters')
plt.ylabel('Sum of squared distance')
plt.title('Elbow method')
plt.show()
```
