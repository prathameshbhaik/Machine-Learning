```python
q1 = data['platelets'].quantile(0.25)
q3 = data['platelets'].quantile(0.75)

iqr = q3-q1
iqr
```

```python
data_iqr = data[~((data['platelets'] < (q1-1.5*iqr))|(data['platelets'] > (q3+1.5*iqr)))]
```

```python
sns.boxplot(x='DEATH_EVENT',y='platelets',data=data_iqr)
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(data_iqr.corr(),annot=True)
plt.show()
```

```python
data_iqr.corr()['DEATH_EVENT'].sort_values(ascending=False)
```

```python
x = data_iqr.drop('DEATH_EVENT',axis=1)
x.shape
```

```python
y = data_iqr['DEATH_EVENT']
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=47)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()

x_train = ss.fit_transform(x_train)
x_test = ss.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
logit = LogisticRegression()
logit.fit(x_train,y_train)
y_pred = logit.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
ac = accuracy_score(y_test,y_pred)
ac
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import roc_curve
y_pred_prob = logit.predict_proba(x_test)[:,1]
fpr,tpr,thresholds = roc_curve(y_test,y_pred_prob)
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr,tpr,label='Logistic Regression')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Logistic Regression ROC Curve')
plt.show()
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test,y_pred_prob)
```
